```python
import pandas as pd
import numpy as np

from surprise import Dataset
from surprise import Reader
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise.accuracy import rmse
```

```python
movies = pd.read_csv("../dataset/movielens/movies.csv")
ratings = pd.read_csv("../dataset/movielens/ratings.csv")
```

```python
reader = Reader(rating_scale=(0.5, 5))
```

```python
data = Dataset.load_from_df(
    ratings[['userId', 'movieId', 'rating']],
    reader
)
```

```python
trainset, testset = train_test_split(
    data,
    test_size=0.2,
    random_state=42
)
```

```python
model = SVD()
```

```python
model.fit(trainset)
```

```python
predictions = model.test(testset)
```

```python
rmse(predictions)
```

```python
model.predict(
    uid=1,
    iid=1
)
```

```python
print(model.predict(1, 50))
```

```python
print(model.predict(5, 200))
```

```python
predictions = model.test(testset)
rmse(predictions)
```

## Conclusion

- Built a Collaborative Filtering recommendation model using the MovieLens dataset.
- Trained an SVD model to learn latent features from user ratings.
- Evaluated the model using RMSE.
- Predicted ratings for unseen user–movie pairs.
