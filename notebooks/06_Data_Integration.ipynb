```python
import pandas as pd
import numpy as np
```

```python
# TMDB Dataset
tmdb_movies = pd.read_csv("../dataset/tmdb/tmdb_5000_movies.csv")

# MovieLens Dataset
ml_movies = pd.read_csv("../dataset/movielens/movies.csv")
ratings = pd.read_csv("../dataset/movielens/ratings.csv")
links = pd.read_csv("../dataset/movielens/links.csv")
```

```python
print(tmdb_movies.head())
print(ml_movies.head())
print(ratings.head())
print(links.head())
```

```python
print("TMDB :", tmdb_movies.shape)
print("MovieLens Movies :", ml_movies.shape)
print("Ratings :", ratings.shape)
print("Links :", links.shape)
```

```python
print(links.head(10))
```

```python
print(tmdb_movies.dtypes)
print(links.dtypes)
```

```python
tmdb_movies.rename(columns={'id':'tmdbId'}, inplace=True)
```

```python
tmdb_movies.columns
```

```python
merged = pd.merge(
    tmdb_movies,
    links,
    on="tmdbId",
    how="inner"
)
```

```python
print(merged.head())
print(merged.shape)
```

```python
final_df = pd.merge(
    merged,
    ratings,
    on="movieId",
    how="left"
)
```

```python
print(final_df.head())
print(final_df.info())
```
